## Data Cleaning

Checking null values, adjusting data types, adding new neccessary columns

```python
#Importing Library
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
#Importing File
df = pd.read_csv('Data/online_retail.csv')
df.head()
```

```python
#Check number of rows
df.shape
```

```python
#Check data types
df.dtypes
```

```python
df.columns
```

```python
#Removing Index Column
df = df[[#index, 
        'InvoiceNo', 'StockCode', 'Description', 'Quantity',
       'InvoiceDate', 'UnitPrice', 'CustomerID', 'Country']].copy()
df.head()
```

```python
#Check if there's null value
df.isna().sum()
```

```python
#Check the null value for description
df.query('Description.isnull()')
```

```python
#Checking the null value for the CustomerID
df.query('CustomerID.isnull()')
```

```python
#Deleting the null value 
df.dropna(subset=['CustomerID','Description'], inplace=True)
df.isna().sum()
```

```python
#Removing the Comma on CustomerID
df['CustomerID'] = df['CustomerID'].astype(int)
#Change date type for InvoiceDate (removing timestamp)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate']).dt.date

df.head()
```

```python
#Add new column for sales 
#Sales = UnitPrice * Quantity 

df['Sales'] = df['UnitPrice'] * df['Quantity']
df.head()
```

```python
#Add New Column for canceled order
#If the number is prefixed with 'c' on column Invoice No, it indicates a cancellation.

df['Status'] = np.where(df.InvoiceNo.str.contains("C"), "Canceled", "Completed")

#Check
df[df.Status == 'Canceled']
```

```python
#Renaming Col Desc to Products
df.rename(columns= {'Description':'Products'},inplace=True)
```

```python
#Converting Cleaned Data To Excel 

file_name = 'cleaned_data.xlsx'
df.to_excel(file_name)
```

```python
#Converting Cleaned Data To CSV
df.to_csv(r'Data/clean_data.csv', index=False)
```
